```python
print("prime number code")
a=int(input())
if a <= 1:
    print(a,"is not a prime number")
else:
    is_prime = True
    for i in range(2, int(a*0.5)+ 1):
        if a % i==0:
            is_prime = False
            break
    if is_prime:
        print(a,"is a prime number")
    else:
        print (a,"is not a prime number")
```
